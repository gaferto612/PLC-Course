## Rung 1 — `Y = A AND B`

**IL:**
```
LD A
AND B
ST Y
```

**ST:**
```pascal
Y := A AND B;
```

## Rung 2 — `Y = (A OR B) AND NOT C`

**IL:**
```
LD A
OR B
ANDN C
ST Y
```

**ST:**
```pascal
Y := (A OR B) AND NOT C;
```
