## System under analysis

A pumped water-tank system with:
- Inlet valve (motorised, fail-closed)
- Pump motor with VFD
- Two level switches (low and high) and one analog level transmitter
- High-temperature switch on the pump motor housing
- E-stop pushbutton (NC, dual-channel)

## Task

Identify at least **10** failure modes. For each, score severity (S), occurrence (O), and detectability (D) on 1–10 scales and compute the Risk Priority Number (RPN = S × O × D). Propose a mitigation for each, then re-score with the mitigation applied.

```python
import pandas as pd

rows = [
    # component, failure mode, effect, S, O, D, mitigation, S', O', D'
    ['Inlet valve', 'Stuck open', 'Tank overflows', 9, 2, 4, 'High-level switch trips pump and alarm', 9, 2, 1],
    ['Inlet valve', 'Stuck closed', 'Pump runs dry', 7, 2, 5, 'Low-level switch interlocks pump', 7, 2, 2],
    ['Level transmitter', 'Wire break (4 mA → 0 mA)', 'False low reading', 6, 3, 3, 'Detect <3.5 mA as fault', 6, 3, 1],
    ['Level transmitter', 'Drift +20 %', 'Tank under-fills or overflows', 6, 4, 7, 'Cross-check with high/low switches', 6, 4, 3],
    ['Pump motor', 'Bearing seizure', 'Smoke, fire risk', 9, 2, 5, 'Thermal overload + temp switch', 8, 2, 2],
    ['VFD', 'Drive trip', 'Pump stops mid-cycle', 5, 3, 2, 'Run-feedback into PLC', 5, 3, 1],
    ['E-stop', 'Single-channel weld', 'E-stop ineffective', 10, 1, 8, 'Dual-channel safety relay', 10, 1, 2],
    ['Low-level switch', 'Stuck active (sees water)', 'Pump runs dry', 7, 2, 6, 'Diagnostic — compare with TX', 7, 2, 3],
    ['High-level switch', 'Stuck inactive', 'Tank overflow risk', 9, 2, 7, 'Diagnostic + redundant overflow weir', 9, 2, 3],
    ['PLC scan', 'Long scan misses E-stop edge', 'Dangerous dwell after stop', 8, 1, 6, 'E-stop hardwired into safety relay', 8, 1, 1],
]

cols = ['Component', 'Failure mode', 'Effect', 'S', 'O', 'D', 'Mitigation', "S'", "O'", "D'"]
df = pd.DataFrame(rows, columns=cols)
df['RPN'] = df['S'] * df['O'] * df['D']
df["RPN'"] = df["S'"] * df["O'"] * df["D'"]
df.sort_values('RPN', ascending=False)
```
