## Specs to translate

1. **Heater on** when (TempLow AND PowerOn) AND NOT Override.
2. **Alarm** when (Door1 OR Door2) is open AND System is armed.
3. **Pump runs** when StartPB is pressed and continues until StopPB is pressed (seal-in latch).
4. **Fault lamp blinks** when ANY of three thermistors is over-temp OR a fuse has blown.
5. **Conveyor advances** when (PartPresent AND Index) AND NOT (Jam OR EStop).
6. **Mixer ready** when Tank1Full AND Tank2Full AND Lid AND NOT Override.
7. **Reject solenoid pulses** when (Sensor_Low AND Sensor_High) — same scan only.
8. **Permissive** when ALL of {GuardClosed, AirOK, OilOK} are true.
9. **Out-of-tolerance** when Pressure < 5 bar OR Pressure > 9 bar.
10. **One-shot** rising edge of StartPB, regardless of how long it stays high.

```python
import itertools

def truth_table(inputs, fn):
    rows = []
    for combo in itertools.product([False, True], repeat=len(inputs)):
        env = dict(zip(inputs, combo))
        rows.append((*combo, fn(**env)))
    header = ' | '.join(inputs + ['Y'])
    print(header)
    print('-' * len(header))
    for row in rows:
        print(' | '.join(str(int(x)) for x in row))

# Example: spec 1 — Heater
truth_table(
    ['TempLow', 'PowerOn', 'Override'],
    lambda TempLow, PowerOn, Override: TempLow and PowerOn and not Override,
)
```
