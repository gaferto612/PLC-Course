````python
import pandas as pd

values = [0, 1, 7, 9, 10, 15, 16, 100, 173, 255, 1024, 4095]

def to_bcd(n):
    return ' '.join(format(int(d), '04b') for d in str(n))

rows = []
for n in values:
    rows.append({
        'denary': n,
        'binary': format(n, 'b'),
        'octal': format(n, 'o'),
        'hex': format(n, 'X'),
        'BCD': to_bcd(n),
    })

pd.DataFrame(rows)
````

## 2. Two's-complement subtraction by hand

Compute `42 - 17` in 8-bit two's complement, by hand, and show your working below.

1. Convert 17 to binary, then take the 2's complement (the 8-bit pattern for `-17`).
2. Add to 42 in binary.
3. Discard any carry-out beyond bit 7. The remaining 8 bits are your result.

**Hand answer:**

```
  +42 = 0010 1010
  -17 = 1110 1111   (= ~0001 0001 + 1)
  ─────────────
        ?           ← write your answer here
```

````python
def twos_complement(n, bits=8):
    """Return the unsigned bit pattern that represents n (which may be negative) in `bits` bits of two's complement."""
    if n < 0:
        n = (1 << bits) + n
    return format(n & ((1 << bits) - 1), f'0{bits}b')

a = 42
b = -17
result = (a + b) & 0xFF
print(f'42  = {twos_complement(a)}')
print(f'-17 = {twos_complement(b)}')
print(f'sum = {twos_complement(result)} = {result if result < 128 else result - 256}')
````
